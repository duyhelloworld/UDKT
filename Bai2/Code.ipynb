```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Đọc dữ liệu từ file Excel
excel_file = './DuLieuEXCEL.xlsx'
df_sinh_vien = pd.read_excel(excel_file, sheet_name='Sinh Vien')
df_diem_mon_1 = pd.read_excel(excel_file, sheet_name='Diem Mon 1')
df_diem_mon_2 = pd.read_excel(excel_file, sheet_name='Diem Mon 2')
df_diem_mon_3 = pd.read_excel(excel_file, sheet_name='Diem Mon 3')
```

```python
# Hiển thị dữ liệu của các DataFrame
print("DataFrame Sinh Vien:")
print(df_sinh_vien)
```

```python
print("\nDataFrame Diem Mon 1:")
print(df_diem_mon_1)
```

```python

print("\nDataFrame Diem Mon 2:")
print(df_diem_mon_2)
```

```python
print("\nDataFrame Diem Mon 3:")
print(df_diem_mon_3)
```

# C

```python
#   Tạo DF mới với cột Diem Tong Ket
df_tong_hop = pd.DataFrame(df_sinh_vien["Ma Sinh Vien"])
df_tong_hop["Diem Tong Ket Mon 1"] = df_diem_mon_1["Diem Qua Trinh"] * 0.3 + df_diem_mon_1["Diem Cuoi Ky"] * 0.7
df_tong_hop["Diem Tong Ket Mon 2"] = df_diem_mon_2["Diem Qua Trinh"] * 0.3 + df_diem_mon_2["Diem Cuoi Ky"] * 0.7
df_tong_hop["Diem Tong Ket Mon 3"] = df_diem_mon_3["Diem Qua Trinh"] * 0.3 + df_diem_mon_3["Diem Cuoi Ky"] * 0.7
```

```python
# Tự quy định hệ số môn 
he_so_mon_1 = 1
he_so_mon_2 = 1
he_so_mon_3 = 1
```

```python
df_tong_hop
```

```python
# Thêm cột Điểm trung bình
df_tong_hop["Diem Trung Binh"] = round((
    df_tong_hop["Diem Tong Ket Mon 1"] * he_so_mon_1
    + df_tong_hop["Diem Tong Ket Mon 2"] * he_so_mon_2
    + df_tong_hop["Diem Tong Ket Mon 3"] * he_so_mon_3
) / (he_so_mon_1 + he_so_mon_2 + he_so_mon_3), 2)
df_tong_hop
```

```python
# Hợp nhất dataframe
df_tong_hop = pd.merge(df_sinh_vien, df_tong_hop, on="Ma Sinh Vien")
df_tong_hop
```

# D

```python
# Lọc ra những sinh viên không qua môn theo từng môn
df_tong_hop[
    (df_tong_hop["Diem Tong Ket Mon 1"] < 4)
    | (df_tong_hop["Diem Tong Ket Mon 2"] < 4)
    | (df_tong_hop["Diem Tong Ket Mon 3"] < 4)
]
```

# E

```python
# Những sinh viên không trượt môn nào 
df_tong_hop[
    (df_tong_hop["Diem Tong Ket Mon 1"] >= 4)
    & (df_tong_hop["Diem Tong Ket Mon 2"] >= 4)
    & (df_tong_hop["Diem Tong Ket Mon 3"] >= 4)
]
```

# F

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Ma Sinh Vien", y="Diem Trung Binh", data=df_tong_hop)
plt.xlabel("Mã sinh viên")
plt.ylabel("Điểm trung bình")
plt.title("Biểu đồ kết quả học tập của sinh viên")
plt.tight_layout()
plt.yticks(np.arange(0, 11, 1))
plt.show()
```

#### lí do chọn biểu đồ cột:
- Thể hiện rõ ràng kết quả học tập theo từng môn học
- Dễ dàng đánh giá được trung bình điểm các sinh viên

#### lí do chọn 2 cột MaSinhVien và DiemTrungBinh
- Cột MaSinhVien là định danh duy nhất cho sinh viên (thay vì HoTen có thể trùng và biểu diễn dài)
- Cột DiemTrungBinh là điểm trung bình các môn của sinh viên
